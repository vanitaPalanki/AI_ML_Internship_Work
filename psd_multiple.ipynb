Power spectral density (PSD) of multiple signals.


```python
import scipy.signal
import librosa
import os
import matplotlib.pyplot as plt

files = os.listdir('/content/files')


for i in enumerate(files):

  path = '/content/files/' + i[1]

  data = librosa.load(path, duration = 5.0, mono = True)
  f,s = scipy.signal.welch(data[0], fs = 10000, nperseg = 4*1024)

  print(len(f))

  print(len(s))

  F = plt.figure()
  F.set_figwidth(10)
  F.set_figheight(10)
  plt.subplot(len(files),1,i[0]+1)
  plt.semilogy(f, s)
  # plt.xlim([0, 100])
  plt.xlabel('frequency')
  plt.ylabel('PSD')
  plt.show()
```
