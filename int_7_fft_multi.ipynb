FFT for multiple audio files.

```python
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np
import librosa

signals_array  = ('BabyElephantWalk60.wav', 'CantinaBand3.wav', 'CantinaBand60.wav', 'Fanfare60.wav', 'ImperialMarch60.wav')

length = len(signals_array)

for i in range(length):

  path = '/content/files/' + signals_array[i]

  data = librosa.load(path, duration=5.0, mono=True)
  N = len(data[0])

  yf = rfft(data[0])
  xf = rfftfreq(N, 1/data[1])

  f = plt.figure()
  f.set_figwidth(10)
  f.set_figheight(10)

  plt.subplot(length, 1, i+1)
  plt.semilogy(xf, np.abs(yf))
  plt.title(" FFT of audio signals ", fontsize = 8, fontweight="bold")
  plt.xlabel("Frequency", fontsize=8)
  plt.ylabel("Magnitude", fontsize=8)
  plt.show
  plt.subplots_adjust(wspace=0.8, hspace = 1.5)


```
