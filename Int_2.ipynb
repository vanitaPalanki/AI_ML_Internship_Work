Import multiple audio files in a loop calculate the mean and plot the graph

```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

audio_files = ('BabyElephantWalk60.wav', 'CantinaBand3.wav', 'CantinaBand60.wav', 'Fanfare60.wav', 'gettysburg.wav', 'gettysburg10.wav', 'ImperialMarch60.wav', 'PinkPanther60.wav', 'preamble.wav', 'StarWars3.wav', 'StarWars60.wav', 'taunt.wav')

length = len(audio_files)

for i in range(length):

  samplerate, data = wavfile.read('/content/files/' + audio_files[i])
  mean = np.mean(data)
  print('mean of data%d:' % i + " " +'%d' % mean)

  plt.plot(data)
  plt.ylabel('Audio signal %d' % i)
  plt.show()
```
