Import an audio signal, plot the original signal in a graph, then change the samplerate to 1000, perform low pass filter and high pass filter with cutoff = 100 and order = 4  . Plot the graph after each change.

```python
from pydub import AudioSegment
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
from scipy import signal

samplerate, data = wavfile.read('/content/files/BabyElephantWalk60.wav')

plt.plot(data)
plt.title("Original audio signal", fontsize=8, fontweight="bold")
plt.xlabel("Samples", fontsize=6)
plt.ylabel('Amplitude', fontsize=6)
plt.show()

input_audio = AudioSegment.from_file('/content/files/BabyElephantWalk60.wav')

desired_sample_rate = 1000

resample_audio = input_audio.set_frame_rate(desired_sample_rate)

resample_audio.export('new_samp.wav', format='wav')

samplerate1, data1 = wavfile.read('new_samp.wav')

plt.plot(data1)
plt.title("Audio signal changed sample rate", fontsize=8, fontweight="bold")
plt.xlabel("Samples", fontsize=6)
plt.ylabel('Amplitude', fontsize=6)
plt.show()

fs = desired_sample_rate/2

b, a = signal.butter(4, 100/fs, btype='lowpass', analog = False)
output = signal.filtfilt(b, a, data1)


plt.plot(output)
plt.title("Audio signal changed after low pass filter", fontsize=8, fontweight="bold")
plt.xlabel("Samples", fontsize=6)
plt.ylabel('Amplitude', fontsize=6)
plt.show()


b, a = signal.butter(4, 100/fs, btype = 'highpass', analog = False)
output_signal = signal.filtfilt(b, a, data1)

plt.plot(output_signal)
plt.title("Audio signal changed after high pass filter", fontsize=8, fontweight="bold")
plt.xlabel("Samples", fontsize=6)
plt.ylabel('Amplitude', fontsize=6)
plt.show()



```
