Import multiple .wav files then in a loop read the data, change its samplerate to 11028, calculate the mean and standard deviation.

```python
import numpy as np
import librosa
from scipy.io import wavfile
import matplotlib.pyplot as plt

file = ['CantinaBand3.wav', 'Fanfare60.wav', 'gettysburg.wav', 'BabyElephantWalk60.wav', 'CantinaBand60.wav']

for i in  enumerate(file):

  path = '/content/files/' + i[1]
  data=  librosa.load(path, duration = 5.0,  mono = True)
  sampled_data = librosa.load(path, sr = 11028, duration = 5.0,  mono = True)
  mean = str(np.mean(data[0]))
  std_dev = str(np.std(data[0]))
  re_mean = str(np.mean(sampled_data[0]))
  re_std_dev = str(np.std(sampled_data[0]))

  f = plt.figure()
  f.set_figwidth(15)
  f.set_figheight(15)

  plt.subplot(len(file),2, i[0]+i[0]+1)
  plt.title("Original signal", fontweight="bold", fontsize= 7)
  plt.xlabel("Frequency", fontsize = 6)
  plt.ylabel("Amplitude", fontsize = 6)
  plt.text(0, 0.2, 'Mean ='+ mean, fontsize= 5)
  plt.text(0, 0.1, 'Std_dev ='+ std_dev, fontsize= 5)
  plt.plot(data[0])
  plt.subplots_adjust(wspace=0.5, hspace=2.0)


  plt.subplot(len(file), 2, i[0]+i[0]+2)
  plt.title("Resampled signal", fontweight="bold", fontsize= 7)
  plt.xlabel("Frequency", fontsize = 6)
  plt.ylabel("Amplitude", fontsize = 6)
  plt.text(0, 0.2, 'Mean ='+ re_mean, fontsize= 5)
  plt.text(0, 0.1, 'Std_dev ='+ re_std_dev, fontsize= 5)
  plt.plot(sampled_data[0])

for i in  enumerate(file):
  path = '/content/files/' + i[1]
  data, fs =  librosa.load(path, duration = 5.0,  mono = True)
  sampled_data, fs1 = librosa.load(path, sr = 11028, duration = 5.0,  mono = True)
  mean = str(np.mean(data[0]))
  std_dev = str(np.std(data[0]))
  re_mean = str(np.mean(sampled_data[0]))
  re_std_dev = str(np.std(sampled_data[0]))
  length = len(data)
  s_length = len(sampled_data)
  t = np.linspace(0, length/fs, length)
  s_t = np.linspace(0, s_length/fs1, s_length)

  f = plt.figure()
  f.set_figwidth(15)
  f.set_figheight(15)

  plt.subplot(len(file),2, i[0]+i[0]+1)
  plt.title("Original signal", fontweight="bold", fontsize= 7)
  plt.xlabel("Time", fontsize = 6)
  plt.ylabel("Amplitude", fontsize = 6)
  plt.text(0, 0.2, 'Mean ='+ mean, fontsize= 5)
  plt.text(0, 0.1, 'Std_dev ='+ std_dev, fontsize= 5)
  plt.plot(t, data)
  plt.subplots_adjust(wspace=0.5, hspace=2.0)


  plt.subplot(len(file), 2, i[0]+i[0]+2)
  plt.title("Resampled signal", fontweight="bold", fontsize= 7)
  plt.xlabel("Time", fontsize = 6)
  plt.ylabel("Amplitude", fontsize = 6)
  plt.text(0, 0.2, 'Mean ='+ re_mean, fontsize= 5)
  plt.text(0, 0.1, 'Std_dev ='+ re_std_dev, fontsize= 5)
  plt.plot(s_t, sampled_data)








```
