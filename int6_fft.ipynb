Import audio file and perform fast Fourier transform

```python
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
import numpy as np
import librosa
from scipy.fft import rfft, rfftfreq

data_music = librosa.load("/content/files/BabyElephantWalk60.wav", duration=5.0, mono = True)
data_voice = librosa.load("/content/files/gettysburg10.wav", duration=5.0, mono = True)

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)

plt.subplot(2,2,1)
plt.title("Original music signal", fontsize=8, fontweight="bold")
plt.xlabel("Frequency", fontsize=6)
plt.ylabel("Magnitute", fontsize=6)
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.plot(data_music[0])

plt.subplot(2,2,2)
plt.title("Original voice signal", fontsize=8, fontweight="bold")
plt.xlabel("Frequency", fontsize=6)
plt.ylabel("Magnitute", fontsize=6)
plt.plot(data_voice[0])


N1 = len(data_music[0])


yf = rfft(data_music[0])
xf = rfftfreq(N1, 1/data_music[1])

plt.subplot(2,2,3)
plt.title("Music signal FFT", fontsize=8, fontweight="bold")
plt.xlabel("Frequency", fontsize=6)
plt.ylabel("Magnitute", fontsize=6)
plt.plot(xf, np.abs(yf))


N2= len(data_voice[0])

yf = rfft(data_voice[0])
xf = rfftfreq(N2, 1/data_voice[1])

plt.subplot(2,2,4)
plt.title("Voice signal FFT", fontsize=8, fontweight="bold")
plt.xlabel("Frequency", fontsize=6)
plt.ylabel("Magnitute", fontsize=6)
plt.plot(xf, np.abs(yf))
plt.show()
```
